Through data exploration, trend analysis, and modeling, I have developed a comprehensive understanding of unemployment patterns, contributing to informed policy decisions and economic forecasting.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('dataset/Unemployment in India.csv')
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Remove rows with missing values
df.dropna(inplace=True)
len(df)
```

```python
# Summary statistics
df.describe()
```

```python
#info about data
df.info()
```

```python
# Unique values in the 'Region' and 'Area' columns
print(df['Region'].unique())
print(df['Area'].unique())
```

```python
# Convert ' Date' column to datetime with the specified format and strip leading spaces
df[' Date'] = pd.to_datetime(df[' Date'].str.strip(), format='%d-%m-%Y')

# Create the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=df, hue='Region')

# Move the legend to the upper right corner
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))

# Add labels and titles
plt.title('Unemployment Rate Over Time')
plt.xlabel(' Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Rename the column to a simpler name
df.rename(columns={' Estimated Unemployment Rate (%)': 'Unemployment Rate'}, inplace=True)

# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Unemployment Rate', data=df)
plt.title('Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Visualize the Distribution of the Unemployment Rate

plt.figure(figsize=(10, 6))
sns.histplot(df['Unemployment Rate'], bins=20, kde=True)
plt.title('Distribution of Unemployment Rate')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
#bar chart to display unemployment rates by region:
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Unemployment Rate', data=df)
plt.title('Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmaps: Explore Correlations Between Different Regions or Variables
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

In conclusion, this analysis of the unemployment rate dataset has revealed several key insights. I found that the region with the highest unemployment rate is Tripura. Conversely, Maharastra has the lowest unemployment rate.
